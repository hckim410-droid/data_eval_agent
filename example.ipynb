```python
####  DB LLM    -- Ollama  
import requests

def call_local_llm(prompt: str,
                   model: str = "llama3.1:8b",
                   temperature: float = 0.2) -> str:
    """
    Ollama  (: http://localhost:11434)  
    Llama3.1 8B    .
    """
    url = "http://localhost:11434/api/generate"

    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False,   #      
        "options": {
            "temperature": temperature,
        }
    }

    resp = requests.post(url, json=payload, timeout=120)
    resp.raise_for_status()
    data = resp.json()

    # Ollama  : {"model": ..., "created_at": ..., "response": "...", ...}
    return data.get("response", "")
```

```python
import duckdb
import pandas as pd
import requests

# ============================================================
# 1. 
# ============================================================

# DuckDB  (telco_poc.duckdb  6   )
con = duckdb.connect("telco_poc.duckdb")

#  LLM(Ollama)  
LOCAL_LLM_MODEL = "llama3.1:8b"
OLLAMA_URL = "http://localhost:11434/api/generate"
#      
OLLAMA_OPTIONS = {
    "temperature": 0.2,
    "top_p": 0.9,
    "num_predict": 800,
}
```

```python
# ============================================================
# 2.    ( DB)
# ============================================================

SCHEMA_BLOCK = """
You are helping with analysis on a telco customer relational database.

We have the following tables and columns:

1) customers
- "Customer ID" (string, primary key)  -> customer_id
- Age (int)
- Gender (string)
- State (string)
- City (string)
- "Zip Code" (string)
- "Tenure in Months" (int)
- Contract (string)                 -- current contract type
- "Payment Method" (string)         -- current payment method
- "Avg Monthly GB Download" (float)
- "Avg Monthly Long Distance Charges" (float)
- "Satisfaction Score" (int)
- CLTV (float)
- Churn (int, 1=churned, 0=active)
- "Customer Status" (string)

2) churn_history_v2
- event_id (string, primary key)
- customer_id (string)
- event_seq (int, 1~N, time order)
- event_month (date)
- status (string: JOINED/STAYED/CHURNED/REJOINED)
- churn_flag (int, 1 if CHURNED, else 0)
- churn_category (string, nullable)
- churn_reason (string, nullable)
- churn_score (float, nullable)

3) internet_service_gaip_history_v3
- contact_id (string)
- customer_id (string)
- event_seq (int)
- event_month (date)
- "Internet Service"       (int or bool)
- "Internet Type"          (string)
- "Online Backup"          (int or bool)
- "Online Security"        (int or bool)
- "Device Protection Plan" (int or bool)
- "Premium Tech Support"   (int or bool)
- "Phone Service"          (int or bool)
- "Multiple Lines"         (int or bool)
- "Streaming Movies"       (int or bool)
- "Streaming Music"        (int or bool)
- "Streaming TV"           (int or bool)
- "Unlimited Data"         (int or bool)

4) usage_history_v1
- contact_id (string)
- customer_id (string)
- event_seq (int)
- event_month (date)
- Monthly_GB_Download (float)
- Monthly_Long_Distance_Charges (float)

5) contract_option_history_v1
- contact_id (string)
- customer_id (string)
- event_seq (int)
- event_month (date)
- Contract (string)
- Payment_Method (string)

6) billing_history_v1
- contact_id (string)
- customer_id (string)
- event_seq (int)
- event_month (date)
- Revenue (float)
- Charge (float)
- Extra_Data_Charges (float)
- Refunds (float)
- Offer (string, nullable)

Keys:
- customers."Customer ID" =   customer_id
-   (customer_id, event_seq)  join 
"""
```

```python
# ============================================================
# 3.     SQL - churn_history_v2 left 
# ============================================================

UNIFIED_TIMELINE_SQL = """
WITH c AS (
  SELECT
    customer_id,
    event_seq AS churn_event_seq,
    CAST(event_month AS DATE) AS event_month,
    status,
    churn_flag,
    churn_category,
    churn_reason,
    churn_score
  FROM churn_history_v2
  WHERE customer_id = ?
),
cust AS (
  SELECT
    customer_id,
    Age,
    Gender,
    State,
    City,
    Zip_Code,
    Tenure_in_Months,
    Contract AS current_contract,
    payment_method,
    Avg_Monthly_GB_Download,
    Avg_Monthly_Long_Distance_Charges,
    Satisfaction_Score,
    CLTV,
    Churn,
    Customer_Status
  FROM customers
),
u AS (
  SELECT
    customer_id,
    CAST(event_month AS DATE) AS event_month,
    Monthly_GB_Download,
    Monthly_Long_Distance_Charges
  FROM usage_history_v1
),
b AS (
  SELECT
    customer_id,
    CAST(event_month AS DATE) AS event_month,
    Revenue,
    Charge,
    Extra_Data_Charges,
    Refunds,
    Offer
  FROM billing_history_v1
),
g AS (
  SELECT
    customer_id,
    CAST(event_month AS DATE) AS event_month,
    Internet_Service,
    Internet_Type,
    Online_Backup,
    Online_Security,
    Device_Protection_Plan,
    Premium_Tech_Support,
    Phone_Service,
    Multiple_Lines,
    Streaming_Movies,
    Streaming_Music,
    Streaming_TV,
    Unlimited_Data
  FROM internet_service_gaip_history_v3
),
co AS (
  SELECT
    customer_id,
    CAST(event_month AS DATE) AS event_month,
    Contract,
    Payment_Method
  FROM contract_option_history_v1
)
SELECT
  c.customer_id,

  -- [customers]  (  )
  cust.Age,
  cust.Gender,
  cust.State,
  cust.City,
  cust.Zip_Code,
  cust.Tenure_in_Months,
  cust.current_contract,
  cust.payment_method,
  cust.Avg_Monthly_GB_Download,
  cust.Avg_Monthly_Long_Distance_Charges,
  cust.Satisfaction_Score,
  cust.CLTV,
  cust.Churn,
  cust.Customer_Status,

  -- [churn_history_v2]  ( )
  c.churn_event_seq AS event_seq,
  c.event_month,
  c.status,
  c.churn_flag,
  c.churn_category,
  c.churn_reason,
  c.churn_score,

  -- [usage_history_v1]
  u.Monthly_GB_Download,
  u.Monthly_Long_Distance_Charges,

  -- [billing_history_v1]
  b.Revenue,
  b.Charge,
  b.Extra_Data_Charges,
  b.Refunds,
  b.Offer,

  -- [internet_service_gaip_history_v3]
  g.Internet_Service,
  g.Internet_Type,
  g.Online_Backup,
  g.Online_Security,
  g.Device_Protection_Plan,
  g.Premium_Tech_Support,
  g.Phone_Service,
  g.Multiple_Lines,
  g.Streaming_Movies,
  g.Streaming_Music,
  g.Streaming_TV,
  g.Unlimited_Data,

  -- [contract_option_history_v1]
  co.Contract,
  co.Payment_Method

FROM c
LEFT JOIN cust ON c.customer_id = cust.customer_id
LEFT JOIN u    ON c.customer_id = u.customer_id  AND c.event_month = u.event_month
LEFT JOIN b    ON c.customer_id = b.customer_id  AND c.event_month = b.event_month
LEFT JOIN g    ON c.customer_id = g.customer_id  AND c.event_month = g.event_month
LEFT JOIN co   ON c.customer_id = co.customer_id AND c.event_month = co.event_month
ORDER BY c.event_month;
"""

def get_unified_timeline(con: duckdb.DuckDBPyConnection,
                         customer_id: str) -> pd.DataFrame:
    """
     customer_id    DataFrame 
    """
    df = con.execute(UNIFIED_TIMELINE_SQL, [customer_id]).df()
    return df
```

```python
# ============================================================
# 4.  DB  
# ============================================================

def build_general_db_prompt(question_ko: str,
                            customer_id: str,
                            timeline_df: pd.DataFrame) -> str:
    """
     DB   LLM   
    """
    if timeline_df.empty:
        table_md = "( customer_id    .)"
    else:
        table_md = timeline_df.to_markdown(index=False)

    prompt = f"""
{SCHEMA_BLOCK}

 customer_id = "{customer_id}"    .
  event_seq(1  ,   ) .

 :
{table_md}

:
{question_ko}

:
-   .
-      ,    .
-  (event_seq) , /// (/ )   .
-   (: Monthly_GB_Download, Revenue )     .
"""
    return prompt.strip()
```

```python
# ============================================================
# 5.  LLM(Ollama)  
# ============================================================

def call_local_llm(prompt: str,
                   model: str = LOCAL_LLM_MODEL,
                   temperature: float = 0.2) -> str:
    """
    Ollama     Llama3.1 8B   .
    """
    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False,
        "options": {
            "temperature": temperature,
        },
    }

    resp = requests.post(OLLAMA_URL, json=payload, timeout=120)
    resp.raise_for_status()
    data = resp.json()
    return data.get("response", "")
```

```python
# ============================================================
# 6. end-to-end  ( LLM )
# ============================================================

def run_one_question_general_db_local(question_ko: str,
                                      customer_id: str,
                                      debug_print: bool = True) -> str:
    """
       ' DB  +  LLM'   end-to-end .
    """
    # 1) DuckDB    
    timeline_df = get_unified_timeline(con, customer_id)

    if debug_print:
        print("===   ( DB ) ===")
        print(timeline_df.head(10))

    # 2)  
    prompt = build_general_db_prompt(
        question_ko=question_ko,
        customer_id=customer_id,
        timeline_df=timeline_df,
    )

    if debug_print:
        print("\n===  LLM   () ===")
        print(prompt[:2000])

    # 3)  LLM 
    answer = call_local_llm(prompt)

    if debug_print:
        print("\n===  LLM  ( DB ) ===")
        print(answer)

    return answer


if __name__ == "__main__":
    #   customer_id 
    test_customer_id = "8634-MPHTR"
    question = (
        "   6    "
    )

    result_local = run_one_question_general_db_local(
        question_ko=question,
        customer_id=test_customer_id,
        debug_print=True,
    )
    ##### DB   ######
```

```python

# 2-2) ont_customer_month : -   
con.execute("""
CREATE OR REPLACE VIEW ont_customer_month AS
WITH c AS (
    SELECT
        customer_id,
        event_seq,
        CAST(event_month AS DATE) AS event_month,
        status,
        churn_flag,
        churn_category,
        churn_reason,
        churn_score
    FROM churn_history_v2
)
SELECT
    -- node_id event_month   (  )
    'CM_' || c.customer_id || '_' || STRFTIME(c.event_month, '%Y%m') AS node_id,

    c.customer_id,
    c.event_seq,          --  (/)
    c.event_month,

    c.status,
    c.churn_flag,
    c.churn_category,
    c.churn_reason,
    c.churn_score,

    g.Internet_Service,
    g.Internet_Type,
    g.Online_Backup,
    g.Online_Security,
    g.Device_Protection_Plan,
    g.Premium_Tech_Support,
    g.Phone_Service,
    g.Multiple_Lines,
    g.Streaming_Movies,
    g.Streaming_Music,
    g.Streaming_TV,
    g.Unlimited_Data,

    co.Contract,
    co.Payment_Method,

    u.Monthly_GB_Download,
    u.Monthly_Long_Distance_Charges,

    b.Revenue,
    b.Charge,
    b.Extra_Data_Charges,
    b.Refunds,
    b.Offer

FROM c
LEFT JOIN usage_history_v1 u
       ON c.customer_id = u.customer_id
      AND c.event_month = CAST(u.event_month AS DATE)

LEFT JOIN billing_history_v1 b
       ON c.customer_id = b.customer_id
      AND c.event_month = CAST(b.event_month AS DATE)

LEFT JOIN internet_service_gaip_history_v3 g
       ON c.customer_id = g.customer_id
      AND c.event_month = CAST(g.event_month AS DATE)

LEFT JOIN contract_option_history_v1 co
       ON c.customer_id = co.customer_id
      AND c.event_month = CAST(co.event_month AS DATE);
""")
```

```python
# ============================================================
# 3. Ontology Node  
#    - ontology_nodes :      
#      node_id / node_type / customer_id / event_seq / properties(JSON)
# ============================================================

# DuckDB JSON (to_json, struct_pack)  
con.execute("""
CREATE OR REPLACE TABLE ontology_nodes AS
-- 3-1) Customer 
SELECT
    node_id,
    'Customer'::VARCHAR    AS node_type,
    customer_id,
    NULL::INTEGER          AS event_seq,
    NULL::DATE             AS event_month,
    to_json(STRUCT_PACK(
        Age                          := Age,
        Gender                       := Gender,
        State                        := State,
        City                         := City,
        Zip_Code                     := Zip_Code,
        Tenure_in_Months             := Tenure_in_Months,
        current_contract             := current_contract,
        current_payment_method       := current_payment_method,
        Avg_Monthly_GB_Download      := Avg_Monthly_GB_Download,
        Avg_Monthly_Long_Distance_Charges := Avg_Monthly_Long_Distance_Charges,
        Satisfaction_Score           := Satisfaction_Score,
        CLTV                         := CLTV,
        Churn                        := Churn,
        Customer_Status              := Customer_Status
    )) AS properties
FROM ont_customer

UNION ALL

-- ======================================================
-- 3-2) CustomerMonth  (   )
-- ======================================================
SELECT
    node_id,
    'CustomerMonth'::VARCHAR AS node_type,
    customer_id,
    event_seq,              -- 
    event_month,             --    
    to_json(STRUCT_PACK(
        event_month                   := event_month,
        status                        := status,
        churn_flag                    := churn_flag,
        churn_category                := churn_category,
        churn_reason                  := churn_reason,
        churn_score                   := churn_score,
        Internet_Service              := Internet_Service,
        Internet_Type                 := Internet_Type,
        Online_Backup                 := Online_Backup,
        Online_Security               := Online_Security,
        Device_Protection_Plan        := Device_Protection_Plan,
        Premium_Tech_Support          := Premium_Tech_Support,
        Phone_Service                 := Phone_Service,
        Multiple_Lines                := Multiple_Lines,
        Streaming_Movies              := Streaming_Movies,
        Streaming_Music               := Streaming_Music,
        Streaming_TV                  := Streaming_TV,
        Unlimited_Data                := Unlimited_Data,
        Contract                      := Contract,
        Payment_Method                := Payment_Method,
        Monthly_GB_Download           := Monthly_GB_Download,
        Monthly_Long_Distance_Charges := Monthly_Long_Distance_Charges,
        Revenue                       := Revenue,
        Charge                        := Charge,
        Extra_Data_Charges            := Extra_Data_Charges,
        Refunds                       := Refunds,
        Offer                         := Offer
    )) AS properties
FROM ont_customer_month

UNION ALL

-- ======================================================
-- 3-3) UsageRecord  ( )
-- ======================================================
SELECT
    node_id,
    'UsageRecord'::VARCHAR  AS node_type,
    customer_id,
    event_seq,
    event_month,
    to_json(STRUCT_PACK(
        event_month                   := event_month,
        Monthly_GB_Download           := Monthly_GB_Download,
        Monthly_Long_Distance_Charges := Monthly_Long_Distance_Charges
    )) AS properties
FROM ont_usage_record

UNION ALL

-- ======================================================
-- 3-4) BillingRecord  ( )
-- ======================================================
SELECT
    node_id,
    'BillingRecord'::VARCHAR AS node_type,
    customer_id,
    event_seq,
    event_month,
    to_json(STRUCT_PACK(
        event_month         := event_month,
        Revenue             := Revenue,
        Charge              := Charge,
        Extra_Data_Charges  := Extra_Data_Charges,
        Refunds             := Refunds,
        Offer               := Offer
    )) AS properties
FROM ont_billing_record

UNION ALL

-- ======================================================
-- 3-5) SubscriptionState  ( / )
-- ======================================================
SELECT
    node_id,
    'SubscriptionState'::VARCHAR AS node_type,
    customer_id,
    event_seq,
    event_month,
    to_json(STRUCT_PACK(
        event_month                   := event_month,
        status                        := status,
        churn_flag                    := churn_flag,
        churn_category                := churn_category,
        churn_reason                  := churn_reason,
        churn_score                   := churn_score,
        Internet_Service              := Internet_Service,
        Internet_Type                 := Internet_Type,
        Online_Backup                 := Online_Backup,
        Online_Security               := Online_Security,
        Device_Protection_Plan        := Device_Protection_Plan,
        Premium_Tech_Support          := Premium_Tech_Support,
        Phone_Service                 := Phone_Service,
        Multiple_Lines                := Multiple_Lines,
        Streaming_Movies              := Streaming_Movies,
        Streaming_Music               := Streaming_Music,
        Streaming_TV                  := Streaming_TV,
        Unlimited_Data                := Unlimited_Data,
        Contract                      := Contract,
        Payment_Method                := Payment_Method
    )) AS properties
FROM ont_subscription_state;
""")

print(" ontology_nodes   ")
```

```python
# ============================================================
# 4. Ontology Edge  
#    - ontology_edges
#    - :
#      Customer --HAS_MONTH_STATE--> CustomerMonth
#      CustomerMonth --HAS_USAGE--> UsageRecord
#      CustomerMonth --HAS_BILLING--> BillingRecord
#      CustomerMonth --HAS_SUBSCRIPTION_STATE--> SubscriptionState
#      CustomerMonth --NEXT--> CustomerMonth( )
# ============================================================

con.execute("""
CREATE OR REPLACE TABLE ontology_edges AS

-- 4-1) Customer -> CustomerMonth
SELECT
    c.node_id AS src_id,
    'HAS_MONTH_STATE'::VARCHAR AS rel_type,
    cm.node_id AS dst_id,
    NULL::INTEGER AS event_seq,
    to_json(STRUCT_PACK(event_month := cm.event_month)) AS properties
FROM ont_customer c
JOIN ont_customer_month cm
  ON c.customer_id = cm.customer_id

UNION ALL

-- 4-2) CustomerMonth -> UsageRecord
SELECT
    cm.node_id AS src_id,
    'HAS_USAGE'::VARCHAR AS rel_type,
    u.node_id AS dst_id,
    NULL::INTEGER AS event_seq,
    to_json(STRUCT_PACK(event_month := cm.event_month)) AS properties
FROM ont_customer_month cm
JOIN ont_usage_record u
  ON cm.customer_id = u.customer_id
 AND CAST(cm.event_month AS DATE) = CAST(u.event_month AS DATE)

UNION ALL

-- 4-3) CustomerMonth -> BillingRecord
SELECT
    cm.node_id AS src_id,
    'HAS_BILLING'::VARCHAR AS rel_type,
    b.node_id AS dst_id,
    NULL::INTEGER AS event_seq,
    to_json(STRUCT_PACK(event_month := cm.event_month)) AS properties
FROM ont_customer_month cm
JOIN ont_billing_record b
  ON cm.customer_id = b.customer_id
 AND CAST(cm.event_month AS DATE) = CAST(b.event_month AS DATE)

UNION ALL

-- 4-4) CustomerMonth -> SubscriptionState
SELECT
    cm.node_id AS src_id,
    'HAS_SUBSCRIPTION_STATE'::VARCHAR AS rel_type,
    s.node_id AS dst_id,
    NULL::INTEGER AS event_seq,
    to_json(STRUCT_PACK(event_month := cm.event_month)) AS properties
FROM ont_customer_month cm
JOIN ont_subscription_state s
  ON cm.customer_id = s.customer_id
 AND CAST(cm.event_month AS DATE) = CAST(s.event_month AS DATE)

UNION ALL

-- 4-5) CustomerMonth -> NEXT ->  CustomerMonth (event_month + 1 month)
SELECT
    cm1.node_id AS src_id,
    'NEXT'::VARCHAR AS rel_type,
    cm2.node_id AS dst_id,
    NULL::INTEGER AS event_seq,
    to_json(STRUCT_PACK(
        from_month := cm1.event_month,
        to_month   := cm2.event_month
    )) AS properties
FROM ont_customer_month cm1
JOIN ont_customer_month cm2
  ON cm1.customer_id = cm2.customer_id
 AND CAST(cm2.event_month AS DATE) = CAST(cm1.event_month AS DATE) + INTERVAL 1 MONTH
;
""")

print(" ontology_edges   ")
```

```python
import duckdb
import pandas as pd
from typing import Tuple, Dict

DB_PATH = "telco_poc.duckdb"
con = duckdb.connect(DB_PATH)

# ============================================================
# 5-1)     (row count / rel_type  / NULL )
# ============================================================

def validate_ontology(con: duckdb.DuckDBPyConnection) -> Dict[str, pd.DataFrame]:
    out = {}

    out["tables"] = con.execute("SHOW TABLES").df()

    out["node_counts_by_type"] = con.execute("""
        SELECT node_type, COUNT(*) AS n
        FROM ontology_nodes
        GROUP BY node_type
        ORDER BY node_type
    """).df()

    out["edge_counts_by_type"] = con.execute("""
        SELECT rel_type, COUNT(*) AS n
        FROM ontology_edges
        GROUP BY rel_type
        ORDER BY rel_type
    """).df()

    # CustomerMonth     ( 7 )
    out["customer_month_stats"] = con.execute("""
        SELECT
          customer_id,
          COUNT(*) AS n_month_nodes,
          MIN(event_month) AS min_month,
          MAX(event_month) AS max_month
        FROM ontology_nodes
        WHERE node_type = 'CustomerMonth'
        GROUP BY customer_id
        ORDER BY n_month_nodes ASC
        LIMIT 20
    """).df()

    # NEXT edge  6 (7)
    out["next_edge_stats"] = con.execute("""
        WITH cm AS (
          SELECT customer_id, COUNT(*) AS n_cm
          FROM ontology_nodes
          WHERE node_type='CustomerMonth'
          GROUP BY customer_id
        ),
        nx AS (
          SELECT
            SUBSTR(src_id, 4, 10) AS customer_id_guess, -- 'CM_'  customer_id      
            COUNT(*) AS n_next
          FROM ontology_edges
          WHERE rel_type='NEXT'
          GROUP BY 1
        )
        SELECT * FROM cm
        ORDER BY n_cm ASC
        LIMIT 20
    """).df()

    return out

checks = validate_ontology(con)
print("=== node_counts_by_type ===")
print(checks["node_counts_by_type"])
print("\n=== edge_counts_by_type ===")
print(checks["edge_counts_by_type"])
print("\n=== customer_month_stats (sample) ===")
print(checks["customer_month_stats"].head(10))
```

```python
# ============================================================
# 5-2)    
#   -  1   /  LLM  
# ============================================================

def get_customer_subgraph(
    con: duckdb.DuckDBPyConnection,
    customer_id: str
) -> Tuple[pd.DataFrame, pd.DataFrame]:
    # 1) customer  +      +  record/state 
    nodes = con.execute("""
        SELECT *
        FROM ontology_nodes
        WHERE customer_id = ?
        ORDER BY
          CASE node_type
            WHEN 'Customer' THEN 0
            WHEN 'CustomerMonth' THEN 1
            WHEN 'SubscriptionState' THEN 2
            WHEN 'UsageRecord' THEN 3
            WHEN 'BillingRecord' THEN 4
            ELSE 9
          END,
          event_month
    """, [customer_id]).df()

    # 2) edges: src/dst         
    #    - node_id  IN   CTE 
    edges = con.execute("""
        WITH my_nodes AS (
          SELECT node_id
          FROM ontology_nodes
          WHERE customer_id = ?
        )
        SELECT e.*
        FROM ontology_edges e
        WHERE e.src_id IN (SELECT node_id FROM my_nodes)
           OR e.dst_id IN (SELECT node_id FROM my_nodes)
        ORDER BY e.rel_type, e.src_id, e.dst_id
    """, [customer_id]).df()

    return nodes, edges
```

```python
import duckdb
import pandas as pd
import requests

# ============================================================
# 1.  ( DB )
# ============================================================

con = duckdb.connect("telco_poc.duckdb")

LOCAL_LLM_MODEL = "llama3.1:8b"
OLLAMA_URL = "http://localhost:11434/api/generate"
OLLAMA_OPTIONS = {
    "temperature": 0.2,
    "top_p": 0.9,
    "num_predict": 800,
}

# ============================================================
# 2.    (Ontology Context)
#    -  DB SCHEMA_BLOCK " "
# ============================================================

SCHEMA_BLOCK_ONTOLOGY_CONTEXT = """
You are helping with analysis on a telco customer ontology-like graph.

You will be given:
- [Ontology Nodes] table: each row is a node with:
  - node_id, node_type, customer_id, event_month, properties(JSON)
- [Ontology Edges] table: each row is a directed relationship with:
  - src_id, rel_type, dst_id, properties(JSON)

Node types:
- Customer: static customer profile node
- CustomerMonth: monthly snapshot node (time slice)
- UsageRecord: monthly usage fact node
- BillingRecord: monthly billing fact node
- SubscriptionState: monthly subscription/options state node

Relationship types:
- HAS_MONTH_STATE: Customer -> CustomerMonth
- HAS_USAGE: CustomerMonth -> UsageRecord
- HAS_BILLING: CustomerMonth -> BillingRecord
- HAS_SUBSCRIPTION_STATE: CustomerMonth -> SubscriptionState
- NEXT: CustomerMonth -> CustomerMonth (next month)

Keys/notes:
- You must answer based ONLY on the provided Ontology Context tables.
- event_month is the time key. Consider time order when explaining changes.
""".strip()

# ============================================================
# 3.    + Ontology Context 
#    (     )
# ============================================================

def get_customer_subgraph(con: duckdb.DuckDBPyConnection, customer_id: str):
    nodes = con.execute("""
        SELECT *
        FROM ontology_nodes
        WHERE customer_id = ?
        ORDER BY
          CASE node_type
            WHEN 'Customer' THEN 0
            WHEN 'CustomerMonth' THEN 1
            WHEN 'SubscriptionState' THEN 2
            WHEN 'UsageRecord' THEN 3
            WHEN 'BillingRecord' THEN 4
            ELSE 9
          END,
          event_month
    """, [customer_id]).df()

    edges = con.execute("""
        WITH my_nodes AS (
          SELECT node_id
          FROM ontology_nodes
          WHERE customer_id = ?
        )
        SELECT e.*
        FROM ontology_edges e
        WHERE e.src_id IN (SELECT node_id FROM my_nodes)
           OR e.dst_id IN (SELECT node_id FROM my_nodes)
        ORDER BY e.rel_type, e.src_id, e.dst_id
    """, [customer_id]).df()

    return nodes, edges


def build_ontology_context(nodes_df: pd.DataFrame, edges_df: pd.DataFrame, max_rows_each: int = 200) -> str:
    #  DB  markdown " "  ( )
    n = nodes_df.copy()
    e = edges_df.copy()

    if "properties" in n.columns:
        n["properties"] = n["properties"].astype(str).str.slice(0, 350)
    if "properties" in e.columns:
        e["properties"] = e["properties"].astype(str).str.slice(0, 350)

    nodes_md = n.head(max_rows_each).to_markdown(index=False)
    edges_md = e.head(max_rows_each).to_markdown(index=False)

    return f"""
[Ontology Nodes]
{nodes_md}

[Ontology Edges]
{edges_md}
""".strip()

#---------------------Ontology CustomerMonth   SQL------------------------------#
def get_ontology_timeline(con, customer_id: str) -> pd.DataFrame:
    sql = """ -- ontology    
WITH cm AS (
  SELECT
    node_id,
    customer_id,
    event_seq,
    CAST(properties->>'event_month' AS DATE) AS event_month,
    properties->>'status' AS status,
    CAST(properties->>'churn_flag' AS INTEGER) AS churn_flag,
    properties->>'churn_category' AS churn_category,
    properties->>'churn_reason' AS churn_reason,
    CAST(properties->>'churn_score' AS DOUBLE) AS churn_score
  FROM ontology_nodes
  WHERE node_type = 'CustomerMonth'
    AND customer_id = ?
),

u AS (
  SELECT
    customer_id,
    CAST(properties->>'event_month' AS DATE) AS event_month,
    CAST(properties->>'Monthly_GB_Download' AS DOUBLE) AS Monthly_GB_Download,
    CAST(properties->>'Monthly_Long_Distance_Charges' AS DOUBLE) AS Monthly_Long_Distance_Charges
  FROM ontology_nodes
  WHERE node_type = 'UsageRecord'
),

b AS (
  SELECT
    customer_id,
    CAST(properties->>'event_month' AS DATE) AS event_month,
    CAST(properties->>'Revenue' AS DOUBLE) AS Revenue,
    CAST(properties->>'Charge' AS DOUBLE) AS Charge,
    CAST(properties->>'Extra_Data_Charges' AS DOUBLE) AS Extra_Data_Charges,
    CAST(properties->>'Refunds' AS DOUBLE) AS Refunds,
    properties->>'Offer' AS Offer
  FROM ontology_nodes
  WHERE node_type = 'BillingRecord'
),

s AS (
  SELECT
    customer_id,
    CAST(properties->>'event_month' AS DATE) AS event_month,
    properties->>'Internet_Service' AS Internet_Service,
    properties->>'Internet_Type' AS Internet_Type,
    properties->>'Online_Backup' AS Online_Backup,
    properties->>'Online_Security' AS Online_Security,
    properties->>'Device_Protection_Plan' AS Device_Protection_Plan,
    properties->>'Premium_Tech_Support' AS Premium_Tech_Support,
    properties->>'Phone_Service' AS Phone_Service,
    properties->>'Multiple_Lines' AS Multiple_Lines,
    properties->>'Streaming_Movies' AS Streaming_Movies,
    properties->>'Streaming_Music' AS Streaming_Music,
    properties->>'Streaming_TV' AS Streaming_TV,
    properties->>'Unlimited_Data' AS Unlimited_Data,
    properties->>'Contract' AS Contract,
    properties->>'Payment_Method' AS Payment_Method
  FROM ontology_nodes
  WHERE node_type = 'SubscriptionState'
)

SELECT
  cm.customer_id,
  cm.event_seq,
  cm.event_month,
  cm.status,
  cm.churn_flag,
  cm.churn_category,
  cm.churn_reason,
  cm.churn_score,

  u.Monthly_GB_Download,
  u.Monthly_Long_Distance_Charges,

  b.Revenue,
  b.Charge,
  b.Extra_Data_Charges,
  b.Refunds,
  b.Offer,

  s.Internet_Service,
  s.Internet_Type,
  s.Online_Backup,
  s.Online_Security,
  s.Device_Protection_Plan,
  s.Premium_Tech_Support,
  s.Phone_Service,
  s.Multiple_Lines,
  s.Streaming_Movies,
  s.Streaming_Music,
  s.Streaming_TV,
  s.Unlimited_Data,
  s.Contract,
  s.Payment_Method

FROM cm
LEFT JOIN u ON cm.customer_id = u.customer_id AND cm.event_month = u.event_month
LEFT JOIN b ON cm.customer_id = b.customer_id AND cm.event_month = b.event_month
LEFT JOIN s ON cm.customer_id = s.customer_id AND cm.event_month = s.event_month
ORDER BY cm.event_month;
 """
    return con.execute(sql, [customer_id]).df()





# ============================================================
# 4. Ontology Context  
#    -  DB  /  
# ============================================================

def build_ontology_prompt(question_ko: str,
                          customer_id: str,
                          ontology_timeline_df: pd.DataFrame,
                          ontology_context: str) -> str:
    if ontology_timeline_df.empty:
        table_md = "(Ontology    .)"
    else:
        table_md = ontology_timeline_df.to_markdown(index=False)

    prompt = f"""
{SCHEMA_BLOCK_ONTOLOGY_CONTEXT}

 customer_id = "{customer_id}"   Ontology Context .
- Nodes/Edges     ///   .
-   event_month  NEXT  .

[Ontology-derived Monthly Timeline]
{table_md}

 :
{ontology_context}

:
{question_ko}

:
-   .
-      ,    .
-  (event_seq  event_month) , /// (/ )   .
-   (: Monthly_GB_Download, Revenue )     .
"""
    return prompt.strip()


# ============================================================
# 5.  LLM(Ollama)   ( DB )
# ============================================================

def call_local_llm(prompt: str,
                   model: str = LOCAL_LLM_MODEL,
                   options: dict = None) -> str:
    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False,
        "options": options or OLLAMA_OPTIONS,
    }
    resp = requests.post(OLLAMA_URL, json=payload, timeout=180)
    resp.raise_for_status()
    data = resp.json()
    return data.get("response", "")


# ============================================================
# 6. end-to-end  (Ontology Context )
# ============================================================

def run_one_question_ontology_context_local(question_ko: str,
                                            customer_id: str,
                                            debug_print: bool = True) -> str:
    # 1)   
    nodes_df, edges_df = get_customer_subgraph(con, customer_id)

    # 2) ontology_context 
    ontology_context = build_ontology_context(nodes_df, edges_df, max_rows_each=200)

    if debug_print:
        print(f"\n--- {customer_id} nodes: {len(nodes_df)}, edges: {len(edges_df)} ---")
        print("\n=== Ontology Context Preview ===")
        print(ontology_context[:8000])

    # 3) Ontology-derived timeline
    timeline_df = get_ontology_timeline(con, customer_id)


    # 4)  
    prompt = build_ontology_prompt(
        question_ko,
        customer_id,
        timeline_df,
        ontology_context
    )

    if debug_print:
        print("\n===  LLM   () ===")
        print(prompt[:5000])

    # 4)  LLM 
    answer = call_local_llm(prompt)

    if debug_print:
        print("\n===  LLM  (Ontology Context) ===")
        print(answer)

    return answer


if __name__ == "__main__":
    test_customer_id = "8634-MPHTR"
    question = "(Contract) (Payment Method)        ?"

    result_local = run_one_question_ontology_context_local(
        question_ko=question,
        customer_id=test_customer_id,
        debug_print=True,
    )
```
